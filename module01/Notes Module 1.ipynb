## Lesson 6 - Data Processing

### Corporate Actions

We usually need to work on adjusted close prices instead of close pirces. Assuming a stock is worth 100 USD today and a 2:1 split (each share is replaced by two shares) is planned at the close, the stock will then be worth 50 USD tomorrow. To adjust for that we could go from stock inception and multiply by n for any n:1 split, this will lead to a today's price very different from where the market is trading, so instead we go backward and divide by n for any n:1 split. Also, **stock tends to rise on stock splits**.

We also need to adjust the price for dividends. **Adjusted price = 1 + D/S**, where D is the dividend and S is the stock price at ex-dividend date. Here again we need to work backwards to ensure today's price matches market trading level.

All these adjusted are usually already included in the adjusted close price.

## Lesson 7 - Stock Returns

### Log returns

For small values of r, ln(r+1) is close to r, and r+1 = p_t / p_(t-1)

Log returns are usually preferred to raw retruns because of their numerical stability (multiply very small values tends to gives results very close to 0, while log values are highers), time additivity, ...

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

x = np.linspace(1.0e-5,1)
y = np.log(x+1)
plt.plot(x,y)
plt.plot(x,x)
plt.title("y=x is tangent to y=ln(x+1) where x is close to 0")
```

If $Y$ is distributed normally, then $e^Y$ is distributed lognormally. Log returns can be considered to IID and normally distributed, so long-term prices and cumulative returns can be modeled as approximately lognormally distributed because they are products of independently, identically distributed (IID) random variables.
